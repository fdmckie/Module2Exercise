#### 1. Find the average score
A program is needed to prompt the user operator for exam scores, calculate the average score and display the result to the screen. The average score is calculated as the sum of the scores divided by the number of scores.

Input: 
	Exam score(s)
	
Process: 
	Keep count of number of exam score entries
	Keep running total of exam scores
	Compute average exam score (total exam score / number of exam score entries)
	
Output: 
	Average exam score




Pseudocode:

BEGIN

  SET COUNTER TO 1
  
  SET TOTAL TO 0
  
	  LOOP UNTIL USER IS FINISHED ENTERING EXAM SCORE(S)
      
		  PROMPT USER FOR EXAM SCORE
          
		  GET EXAM SCORE
          
		  ADD EXAM SCORE TO TOTAL
          
		  ADD 1 TO COUNTER
          
		  ASK USER FOR ANOTHER EXAM SCORE
          
			IF ANOTHER ENTRY, CONTINUE LOOP
            
				IF NOT, GO TO ENDLOOP
                
	  ENDLOOP
	
  AVERAGE = DIVIDE TOTAL BY COUNTER
  
  PRINT AVERAGE
  
END

#### 2. How much water runs off a roof in a rain storm?
To calculate the runoff from any given rainfall: Take the dimensions of the footprint of the roof and convert them to inches. (So, a 50' x 20' roof is 600" x 240"). Multiply the roof dimensions by the number of inches of rainfall. As an example, 600" x 240" x 1" = 144,000 cubic inches of water for an inch of rainfall. Divide that result by 231 to get the number of gallons (because 1 gallon = 231 cubic inches). (144,000/231 = 623.38).

Input: 
	rainfall (in inches)
	dimensions (in 2-D) of the footprint of the roof (in feet)

Process: 
	convert dimensions of the footprint of the roof into inches (multiply dimension measurements by 12)
    
	compute (in cubic inches) amount of rainfall on roof (rainfall x dimensions of footprint of roof (in inches))
    
	convert, into gallons, amounf of rainfall on roof (divide amount of rainfall on roof by 231)

Output: 

	number of gallons of rainfall onto roof


Pseudocode:

BEGIN

	PROMPT USER FOR AMOUNT OF RAINFALL (IN INCHES)
    
	PROMPT USER FOR DIMENSIONS OF ROOF (IN FEET)
	
	CONVERT DIMENSION OF ROOF TO INCHES (x 12 FOR BOTH VALUES)
    
	COMPUTE AMOUNT OF RAINFALL ON ROOF (IN CUBIC INCHES)
    
	CONVERT AMOUNT OF RAINFALL ON ROOF TO GALLONS

END

#### 3. Process customer record
A program is required to read a customer’s name, a purchase amount and a tax code. The tax code has been validated and will be one of the following: 0 tax exempt (0%) 1 state sales tax only (3%) 2 federal and state sales tax (5%) 3 special sales tax (7%) The program must then compute the sales tax and the total amount due,and print the customer’s name, purchase amount, sales tax and total amount due.

Input: 
	Customer's name
    
	Purchase amount
    
	Tax code

Process: 
	Enter customer's name
    
	Enter purchase amount
    
	Enter tax code
    
		Convert tax code into sales tax value
        
	Compute sales tax amount (purchase amount * sales tax value)
    
	Compute total amount due (purchase amount + sales tax amount)
	
Output:

	Print the following information:
    
		Customer name
        
		Purchase amount
        
		Sales tax amount & Sales tax
        
		Total amount due
		
		
Pseudocode: 

BEGIN

	READ CUSTOMER NAME
    
	READ PURCHASE AMOUNT
    
	READ TAX CODE
	
	CONVERT TAX CODE INTO SALES TAX (IN %)
    
		0 - TAX EXEMPT (0%)
        
		1 - STATE SALES TAX ONLY (3%)
        
		2 - FEDERAL & STATE SALES TAX (5%)
        
		3 - SPECIAL SALES TAX (7%)
		
	COMPUTE SALES TAX AMOUNT (PURCHASE AMOUNT x SALES TAX)
    
	COMPUTE TOTAL AMOUNT DUE (PURCHASE AMOUNT + SALES TAX AMOUNT)
	
	PRINT CUSTOMER NAME
    
	PRINT PURCHASE AMOUNT
    
	PRINT SALES TAX AMOUNT AND SALES TAX (IN %)
    
	PRINT TOTAL AMOUNT DUE

END

#### 4. Calculate employee's pay 
A program is required by a company to read an employee’s number, pay rate and the number of hours worked in a week. The program is then to validate the pay rate field and the hours worked field and, if valid, compute the employee’s weekly pay and then print it and the input data.

Validation: According to the company’s rules, the maximum hours an employee can work per week is 60 hours, and the maximum hourly rate is $25.00 per hour. If the hours worked field or the hourly rate field is out of range, the input data and an appropriate message are to be printed and the employee’s weekly pay is not to be calculated.

Weekly pay calculation: Weekly pay is calculated as hours worked times pay rate. If more than 35 hours are worked, payment for the overtime hours worked is calculated at time-and-a-half.

Input: 

	Employee's number
    
	Pay rate
    
	Number of hours worked in a week

Process: 

	Read employee's number
    
	Read pay rate
    
	Read number of hours worked in a week
    
	Once pay rate & number of hours worked in a week are vadilated
    
		compute normal weekly pay of employee
        
		if necessary, compute overtime pay for the week
        
		compute total weekly pay for employee (normal weekly pay + overtime pay)

Output: 

	Print Employee number
    
	Print pay rate
    
	Print regular weekly pay
    
	If necessary, print overtime pay for the week
    
	Print total pay for the week


Pswudocode: 

	BEGIN
	
		READ EMPLOYEE NUMBER
        
		READ PAY RATE
        
		READ HOURS WORKED (FOR 1 WEEK)
        
		
		VALIDATE PAY RATE (LESS THAN OR EQUAL TO 25)
        
		VALIDATE HOURS WORKED (LESS THAN OR EQUAL TO 60)
        
		
		IF PAY RATE OR HOURS WORKED ARE INVALID
        
			PRINT APPROPRIATE MESSAGE DENOTING INVALID DATA
            
			GO TO END OF PROGRAM
            
		END IF
		
		IF HOURS WORKED > 35
        
			COMPUTE OVERTIME HOURS (HOURS WORKED - 35)
            
			COMPUTE OVERTIME PAY RATE (PAY RATE x 1.5)
            
			COMPUTE OVERTIME PAY (OVERTIME HOURS x OVERTIME PAY RATE)
            
			COMPUTE NORMAL PAY (35 x PAY RATE)
            
			COMPUTE TOTAL PAY (NORMAL PAY + OVERTIME PAY)
            
			
			PRINT EMPLOYEE NUMBER
            
			PRINT HOURS WORKED
            
			PRINT OVERTIME HOURS WORKED
            
			PRINT PAY RATE
            
			PRINT OVERTIME PAY RATE
            
			PRINT NORMAL PAY
            
			PRINT OVERTIME PAY
            
			PRINT TOTAL PAY
            
		END IF

		IF HOURS WORKED <= 35
        
			COMPUTE TOTAL PAY (HOURS WORKED x PAY RATE)
			
			PRINT EMPLOYEE NUMBER
            
			PRINT HOURS WORKED
            
			PRINT PAY RATE
            
			PRINT TOTAL PAY
            
		END IF
		
	END
